```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
en=OrdinalEncoder()
```

```python
a=['Class']
df[a]=en.fit_transform(df[a])
```

```python
df=df.astype({'Class':'int','I0':'int','DA':'int','Area':'int','A/DA':'int','Max IP':'int','DR':'int','P':'int'})
```

```python
# Train-Test split.
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,1:10]
y=df['Class']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xtrain
```

```python
ytrain
```

```python
# Model Building.
from sklearn.ensemble import RandomForestClassifier
```

```python
mod=RandomForestClassifier()
```

```python
mod.fit(xtrain,ytrain)
```

```python
pred=mod.predict(xtest)
print(pred)
```

```python
# Accuracy Calculation.
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(ytest,pred)
print("Accuracy is:",acc*100)
```

```python
# Confusion matrix.
from sklearn.metrics import confusion_matrix
```

```python
c=confusion_matrix(ytest,pred)
print("Confusion Matrix is: \n",c)
```

```python
# Plot confusion matrix
sns.heatmap(c,annot=True,fmt='g',cmap='Oranges',xticklabels=['car', 'fad', 'mas', 'gla', 'con', 'adi'],yticklabels=['car', 'fad', 'mas', 'gla', 'con', 'adi'])
plt.xlabel('Actual',fontsize=15)
plt.ylabel('Prediction',fontsize=15)
plt.title('Confusion Matrix',fontsize=20)
plt.show()
```
